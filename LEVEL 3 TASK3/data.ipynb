# Loading and Inspecting the Data

```python
import pandas as pd

# Load the dataset
file_path = 'data/Dataset .csv'  
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
```

# Distribution of Ratings

```python
import matplotlib.pyplot as plt

# Histogram of ratings
plt.figure(figsize=(10, 6))
plt.hist(data['Aggregate rating'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Average Ratings by Cuisines or Cities

# Code Snippet for Average Rating by Cuisines:

```python
# Average rating by cuisine
cuisine_ratings = data.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

# Bar plot for cuisines
plt.figure(figsize=(12, 6))
cuisine_ratings.plot(kind='bar', color='coral')
plt.title('Top 10 Cuisines by Average Rating')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

# Code Snippet for Average Rating by Cities:

```python
# Average rating by city
city_ratings = data.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

# Bar plot for cities
plt.figure(figsize=(12, 6))
city_ratings.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Cities by Average Rating')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

# Relationship Between Features and Target Variable

# Code Snippet for Scatter Plot (Price Range vs Rating):

```python
# Scatter plot for Price Range vs Rating
plt.figure(figsize=(10, 6))
plt.scatter(data['Price range'], data['Aggregate rating'], alpha=0.5, color='purple')
plt.title('Price Range vs Aggregate Rating')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```

# Classify Cities as Urban or Local

```python
# Define urban and local cities
urban_cities = ['New York', 'London', 'Dubai', 'Mumbai', 'Tokyo']
local_cities = ['Pocatello', 'Winchester Bay', 'Yorkton', 'Clatskanie', 'Cochrane']

# Classify each row in the dataset as urban or local
data['Location Type'] = data['City'].apply(lambda x: 'Urban' if x in urban_cities else ('Local' if x in local_cities else 'Other'))

# Display the first few rows to verify
data.head()
```

# Create Visualizations


# Average Rating Comparison

```python
import matplotlib.pyplot as plt

# Calculate average ratings for urban and local restaurants
average_ratings = data.groupby('Location Type')['Aggregate rating'].mean()

# Bar plot for average ratings
average_ratings.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Average Rating Comparison')
plt.xlabel('Location Type')
plt.ylabel('Average Rating')
plt.show()
```

# Count of Restaurants:

```python
# Count of restaurants in each location type
restaurant_counts = data['Location Type'].value_counts()

# Bar plot for count of restaurants
restaurant_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.title('Count of Restaurants by Location Type')
plt.xlabel('Location Type')
plt.ylabel('Number of Restaurants')
plt.show()
```

#  Ratings Analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# Replace 'your_dataset.csv' with the actual path of your dataset
new_dataset = pd.read_csv('data/Dataset .csv')

# Average rating per city
avg_rating_per_city = new_dataset.groupby('City')['Aggregate rating'].mean()
print("Average Rating Per City:\n", avg_rating_per_city)

# Plotting
avg_rating_per_city.plot(kind='bar', figsize=(10, 6))
plt.title('Average Restaurant Rating per City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Cuisine Popularity

```python
# Counting the number of restaurants offering each cuisine
cuisine_popularity = new_dataset['Cuisines'].str.split(', ').explode().value_counts()
print("Cuisine Popularity:\n", cuisine_popularity)

# Plotting
cuisine_popularity.head(10).plot(kind='bar')  # Top 10 cuisines
plt.title('Top 10 Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.show()
```

#  Geographical Analysis

```python
import folium

# Create a base map
map = folium.Map(location=[new_dataset['Latitude'].mean(), new_dataset['Longitude'].mean()], zoom_start=6)

# Add markers for each restaurant
for idx, row in new_dataset.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Restaurant Name']).add_to(map)

# Save the map to an HTML file
map_file_path = 'restaurants_map.html'
map.save(map_file_path)

```

# Price Range Analysis

```python
# Relationship between price range and ratings
price_rating_relation = new_dataset.groupby('Price range')['Aggregate rating'].mean()
print("Price Range and Rating Relation:\n", price_rating_relation)

# Plotting
price_rating_relation.plot(kind='bar')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()

```

# Average Aggregate Rating by Currency

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv('data/Dataset .csv')

# Group the data by currency and calculate the average rating
currency_rating = data.groupby('Currency')['Aggregate rating'].mean()

# Sort the results for better visualization
currency_rating_sorted = currency_rating.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=currency_rating_sorted.index, y=currency_rating_sorted.values)
plt.xticks(rotation=45)
plt.xlabel('Currency')
plt.ylabel('Average Aggregate Rating')
plt.title('Average Aggregate Rating by Currency')
plt.show()
```

# Ratings Based on Locality

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
data = pd.read_csv('data/Dataset .csv')

# Group the data by 'Locality' and calculate the average rating
locality_rating = data.groupby('Locality')['Aggregate rating'].mean().sort_values(ascending=False)

# Display the top localities (you can adjust the number)
top_localities = locality_rating.head(20)

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=top_localities.values, y=top_localities.index)
plt.xlabel('Average Aggregate Rating')
plt.ylabel('Locality')
plt.title('Average Aggregate Rating by Locality')
plt.show()
```

# Correlation Between Latitude/Longitude and Ratings

```python
# Considering only relevant columns
relevant_data = data[['Latitude', 'Longitude', 'Aggregate rating']]

# Calculate the correlation
correlation = relevant_data.corr()

# Plot the correlation matrix
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation between Latitude, Longitude, and Ratings')
plt.show()
```
