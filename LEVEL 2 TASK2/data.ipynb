# Data Cleaning and Preparation

```python
import pandas as pd

# Load the dataset
file_path = 'data/Dataset .csv'
data = pd.read_csv(file_path)

# Checking for missing values in relevant columns
missing_values_check = data[['Price range', 'Aggregate rating', 'Rating color']].isnull().sum()

# Checking data types of these columns
data_types_check = data[['Price range', 'Aggregate rating', 'Rating color']].dtypes

# Checking for any outliers in 'Aggregate rating' (ratings should be within a sensible range, e.g., 0 to 5)
rating_outliers_check = data[(data['Aggregate rating'] < 0) | (data['Aggregate rating'] > 5)]['Aggregate rating'].count()

# Print the results
print("Missing Values Check:\n", missing_values_check)
print("\nData Types Check:\n", data_types_check)
print("\nRating Outliers Check (number of outliers):\n", rating_outliers_check)
```

# Determine the Most Common Price Range

```python
import pandas as pd

# Load the dataset
file_path = 'data/Dataset .csv'
data = pd.read_csv(file_path)

# Calculating the frequency of each price range
price_range_frequency = data['Price range'].value_counts()

# Identifying the most common price range
most_common_price_range = price_range_frequency.idxmax()

# Print the frequency of each price range and the most common price range
print("Frequency of Each Price Range:\n", price_range_frequency)
print("\nMost Common Price Range:", most_common_price_range)
```

# Identify the Highest Average Rating and Corresponding Color

```python
import pandas as pd

# Load the dataset
file_path = 'data/Dataset .csv'
data = pd.read_csv(file_path)

# Check for missing values in relevant columns
missing_values = data[['Price range', 'Aggregate rating', 'Rating color']].isnull().sum()

# Ensuring there are no missing values
if missing_values.sum() == 0:
    # Calculate the average rating for each price range
    average_rating_by_price_range = data.groupby('Price range')['Aggregate rating'].mean()

    # Identify the price range with the highest average rating
    highest_avg_rating_price_range = average_rating_by_price_range.idxmax()

    # Determine the most frequent 'Rating color' for this price range
    color_for_highest_avg_rating = data[data['Price range'] == highest_avg_rating_price_range]['Rating color'].mode()[0]

    # Print the results
    print("No missing values found in the key columns.")
    print("Highest Average Rating is in Price Range:", highest_avg_rating_price_range)
    print("The corresponding color for this rating is:", color_for_highest_avg_rating)
else:
    print("Missing values found in the dataset. Please clean the data before proceeding.")
```

# Creating the scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'data/Dataset .csv'
data = pd.read_csv(file_path)

# Scatter plot of Aggregate Rating vs Price Range, colored by Rating Color
plt.figure(figsize=(10, 6))
for color in data['Rating color'].unique():
    # Filter data for each color
    color_data = data[data['Rating color'] == color]
    plt.scatter(color_data['Price range'], color_data['Aggregate rating'], label=color, alpha=0.7)

plt.title('Aggregate Rating vs Price Range, Colored by Rating Color')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.legend(title='Rating Color')
plt.grid(True)
plt.show()
```

# Statistical Summary and Correlation Analysis

```python
import pandas as pd

# Load the dataset
file_path = 'data/Dataset .csv'
data = pd.read_csv(file_path)

# Display a statistical summary of the dataset
print("Statistical Summary of the Dataset:")
print(data.describe())

# Correlation Analysis
# Let's check if there's any notable correlation between 'Price range' and 'Aggregate rating'
correlation_matrix = data[['Price range', 'Aggregate rating']].corr()
print("\nCorrelation between Price Range and Aggregate Rating:")
print(correlation_matrix)

# Optionally, visualize the correlation matrix using a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# Analyze Average Rating by Cuisine Type Within Each Price Range

```python
import pandas as pd

# Load the dataset
file_path = 'data/Dataset .csv'  
data = pd.read_csv(file_path)

# Splitting the 'Cuisines' column into individual cuisines for each restaurant
# We'll create a new row for each cuisine type for each restaurant
data_expanded = data.drop('Cuisines', axis=1).join(
    data['Cuisines'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True).rename('Cuisine')
)

# Grouping by Price Range and Cuisine, then calculating the average rating
average_rating_by_cuisine = data_expanded.groupby(['Price range', 'Cuisine'])['Aggregate rating'].mean().reset_index()

# Sort the results for better readability
sorted_average_ratings = average_rating_by_cuisine.sort_values(by=['Price range', 'Aggregate rating'], ascending=[True, False])

# Displaying the top 5 cuisines for each price range based on average ratings
for price_range in sorted_average_ratings['Price range'].unique():
    print(f"Top 5 Cuisines in Price Range {price_range} by Average Rating:")
    top_cuisines = sorted_average_ratings[sorted_average_ratings['Price range'] == price_range].head(5)
    print(top_cuisines, "\n")

# Note: This code assumes that the 'Cuisines' column contains a comma-separated list of cuisines for each restaurant.
```

# Stacked Bar Chart for Rating Color Distribution in Each Price Range

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'data/Dataset .csv'
data = pd.read_csv(file_path)

# Counting the frequency of each rating color within each price range
color_distribution = data.groupby(['Price range', 'Rating color']).size().unstack(fill_value=0)

# Plotting a stacked bar chart
color_distribution.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Distribution of Rating Colors in Each Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.legend(title='Rating Color')
plt.xticks(rotation=0)
plt.show()
```
